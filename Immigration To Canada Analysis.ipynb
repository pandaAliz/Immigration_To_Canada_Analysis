```python
file_path = 'data.xlsx'
df_can = pd.read_excel(
    file_path,
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)
print('Data downloaded and read into a dataframe!')
```

```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
```

```python
df_can
```

```python
df_can.rename(columns = {'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'}, inplace = True)

# Sum rows to get total immigration by country (numeric only values)
df_can['Total'] = df_can.sum(axis=1, numeric_only=True)

# Set country name as index
df_can.set_index('Country', inplace=True)
```

```python
df_can
```

```python
# Convert column names to str to avoid consfusing with index
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(2000, 2014)))
```

```python
nepal = df_can.loc['Nepal',years] 
nepal.index = nepal.index.map(int)
nepal.plot(kind = 'line')

plt.title('Immigration from Nepal')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')

plt.show() 
```
